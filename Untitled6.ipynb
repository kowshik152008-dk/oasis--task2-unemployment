```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
print(df.head())
```

```python
df.shape
df.info()
df.isnull().sum()
df.describe()
```

```python
df.columns = df.columns.str.strip()  
df.dropna(inplace=True)              
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)  
```

```python

fig2 = px.line(df, 
               x='Date', 
               y='Estimated Unemployment Rate (%)',
               color='Region', 
               title='Unemployment Rate Over Time')
fig2.show()
```

```python
fig3 = px.scatter_geo(df, 
                      lat='latitude', 
                      lon='longitude',
                      color='Estimated Unemployment Rate (%)',
                      hover_name='Region', 
                      size='Estimated Unemployment Rate (%)',
                      title='Unemployment Rate Map - India',
                      scope='asia')
fig3.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), 
            annot=True, 
            cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
fig4 = px.box(df, 
              x='Region', 
              y='Estimated Unemployment Rate (%)',
              color='Region',
              title='Unemployment Rate Distribution by Region')
fig4.show()
```

```python
print("=== KEY FINDINGS ===")
print(f"Total Records: {df.shape[0]}")
print(f"States/Regions covered: {df['Region'].nunique()}")
print(f"\nHighest Unemployment Rate: {df['Estimated Unemployment Rate (%)'].max():.2f}%")
print(f"Lowest Unemployment Rate: {df['Estimated Unemployment Rate (%)'].min():.2f}%")
print(f"Average Unemployment Rate: {df['Estimated Unemployment Rate (%)'].mean():.2f}%")
```
